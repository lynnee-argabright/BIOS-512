Can you make this plot? üëá What does this plot demonstrate?

![](img/patchwork_example_penguins.png)

üìçHint: add `scale_color_brewer(type = "qual", palette = "Set2")` to adjust the colors of the second plot!

```r
penguin= ggplot(penguins, aes(x=body_mass_g,
                              y=flipper_length_mm,
                              color=species))+
        geom_point(na.rm=TRUE)
       #na.rm=TRUE to remove missing values so you don't get a warning message
penguin
```

```r
penguin_byisland = ggplot(penguins, aes(x=body_mass_g, y=flipper_length_mm, color=island))+
                    scale_color_brewer(type="qual",palette="Set2")+
                    geom_point()
penguin_byisland

#Anytime you want to play with what the chart looks like, it'll be adjusted with a scale_ command
```

```r
options(repr.plot.width=8)
penguin+penguin_byisland
```
