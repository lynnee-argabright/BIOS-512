Now that you've seen all the available geoms, use the line `geom` to make a line chart from the `economics` data set with **`unemploy` on the  y-axis** and **`date` on the x-axis.** 

```r
# chart code goes here...

line = ggplot(economics, aes(x=date, y=unemploy, alpha=pop)) +
        geom_line(color='steelblue') #alpha makes line more transparent the closer to 0 it is. Could put in ggplot or in geom_line
        #geom_line(alpha=0.05)
line
```
