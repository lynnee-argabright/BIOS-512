4. Add an aesthetic mapping for size to sepal width.

```r
options(repr.plot.width=4, repr.plot.height=3)
flower=ggplot(iris, aes(x=Petal.Width, y=Petal.Length, size=Sepal.Width))+
        geom_point()
flower
#size=how big the dots are
#one way to make the dots more apparent is to facet the graph
```

5. Create a facets plot where each facet represents one iris species. 
(You may have to resize the output again.)

```r
options(repr.plot.width=8, repr.plot.height=6)
flower=ggplot(iris, aes(x=Petal.Width, y=Petal.Length, size=Sepal.Width))+
        geom_point(alpha=0.25)+
        #scale_size(range=c(0.1,2.0))+
        facet_wrap(~Species, nrow=2)
flower
#alpha will make dots more transparent so you can better see overlapping data
#scale size changes how far away the dots appear in the chart

#Usually you want to reserve the mappings of more global chart properties like what's on the axes
#as they actually are in ggplot; for more aesthetic changes, put in geom_point. 

#You could alternately do ggplot(iris, aes(x=Petal.Width, y=Petal.Length, alpha=Sepal.Width))
#but that changes what you're asking for
floweralpha=ggplot(iris, aes(x=Petal.Width, y=Petal.Length, alpha=Sepal.Width))+
        geom_point(alpha=0.25)+
        facet_wrap(~Species, nrow=2)
floweralpha
```
